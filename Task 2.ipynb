```python
#Let's plot barcharts among age , sex and heart diseases
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='chest_pain_type', y='age', hue='sex' ,  data=df)  
plt.title('Bar Plot among chest pain type , age and sex')
plt.xlabel('Chest pain type')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='resting_blood_pressure', data=df)
plt.title('Bar Plot of Resting Blood Pressure')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Count')
plt.xticks(rotation=90) 
plt.show()
```
